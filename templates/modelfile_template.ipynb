```python
import json
import builtins
from pathlib import Path

def print_system_tree(json_file="/home/flo/workspace/00_config/paths.json"):
    with builtins.open(json_file, 'r') as f:
        data = json.load(f)

    print(f"\n{'ALIAS / TARGET':<20} | {'PATH':<30} | {'DESCRIPTION'}")
    print("=" * 90)

    # 1. Pfade ausgeben
    print("--- [DIRS] ---")
    for alias, info in data['paths'].items():
        # HIER HAT DIE ZEILE GEFEHLT:
        print(f"{alias:<20} | {info['path']:<30} | {info.get('desc', '')}")

    # 2. Targets ausgeben (jetzt mit klarer Markierung)
    print("\n--- [TARGETS] ---")
    for name, info in data['targets'].items():
        print(f"FILE: {name:<14} | {info['path']:<30} | {info['desc']}")

if __name__ == "__main__":
    print_system_tree()
```
