# global_agent_directives

## architectural_authority
- **Source of Truth:** The `Leleka` system architecture (manifesto in `01_ai_stack/context/leleka_manifesto.md`) defines the global architecture. All other projects must strictly adhere to this architectural structure.
- **Standards:** Coding standards (PEP 8, `uv`, `snake_case`), naming conventions, and safety protocols are **globally binding**. A local `CLAUDE.md` may never loosen these standards, only supplement them with project-specific requirements.

## access_control_and_scopes
- **Global Read-Only:** Access to the entire repository root is permitted if necessary for the task.
- **Local Write-Scope:** Write access is **strictly limited to the current project directory** (where the project's `CLAUDE.md` is located).
- **External Write-Scope:** Write operations outside the current project directory are **strictly prohibited** unless the user grants explicit permission by providing the specific absolute path from the repository root.
- **Protected Files:** `01_ai_stack/context/leleka_manifesto.md`, `CLAUDE.md`, and `TODO.md` are **READ-ONLY**. Any modifications to these must be initiated or explicitly confirmed by the user.
- **Path Handling:** Paths outside the current project must be specified as absolute paths from the repository root (e.g., `01_ai_stack/config/...`). The agent must not guess locations.

## session_cycle
1. **Initial Setup:** Identify current project directory, read local `CLAUDE.md` and `TODO.md`.
2. **Architecture Sync:** Read `01_ai_stack/context/leleka_manifesto.md` (Read-Only) to ensure architectural compliance.
3. **Task & Branch Verification:** 
    a. Check `TODO.md` and ask the user to select or define the task.
    b. Verify current branch (`git branch --show-current`). 
    c. If not on `dev_[project_name]` or `fix_[project_name]`, create/switch branch before performing any file operations.
4. **Execution:** Proceed with task. 
5. **Session End:** Commit all changes locally; never push to remote without explicit user confirmation.

## git_guardrails
- **Git Protection:** Working directly on `main` or `master` is strictly forbidden.
- **Enforcement:** If branch verification fails during the `session_cycle`, the agent must halt all write operations until the correct branch is confirmed.

```python
import subprocess
from pathlib import Path

# --- HIER DEINE DATEINAMEN EINSTELLEN ---
source_notebook = "leleka_core_ctx.ipynb"  # Der Name deiner Datei
target_filename = "CLAUDE"        # Der Name der Zieldatei (ohne .md)
# ----------------------------------------

# Zielpfad im context-Ordner
target_folder = Path("/home/flo")

# Export-Befehl via Jupytext
# Er nimmt die Quelle und schreibt sie explizit in den Zielpfad mit dem neuen Namen
cmd = [
    "jupytext",
    "--to", "markdown",
    "--output", str(target_folder / f"{target_filename}.md"),
    source_notebook
]

try:
    subprocess.run(cmd, check=True)
    print(f"✓ Export erfolgreich! Datei '{source_notebook}' wurde als '{target_filename}.md' exportiert.")
except subprocess.CalledProcessError as e:
    print(f"✗ Fehler beim Export: {e}")
```
