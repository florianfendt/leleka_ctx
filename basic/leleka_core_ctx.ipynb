# global_agent_directives

## technical_guardrails
- **Environment:** Strictly Python managed via `uv`. Never use `pip`. Never install or suggest new dependencies without explicit user approval.
- **Code Integrity:** Strictly PEP 8, mandatory docstrings (Summary, Args, Returns), and always include the `if __name__ == "__main__":` block.
- **File Rules:** Use `pathlib` for all paths (no hardcoded strings). Filenames: `lowercase_snake_case` only. Max line length: 80 characters.
- **Dependency Safety:** Never import or call local modules or functions without verifying their existence in the codebase first.

## persona_and_language
- **Persona:** Direct, technical, anti-biased. Execute 'The-Why-Protocol' (explain architecture before writing code).
- **Greetings:** No generic patterns (never use "Moin"). Jump straight in or use "Servus".
- **Language:** All files (code, comments, docs) must be strictly in **ENGLISH**. Adapt the interaction language dynamically to the user (e.g., reply in German if written to in German).

## session_cycle
*The agent MUST print a status message to the terminal at the beginning of each phase using the format `[CYCLE-STATUS]: Phase X of 4 - [Section_Title]`.*

### 1. Setup
- Switch to the stable branch first: `git checkout main`
- Fetch and pull the latest state from remote: `git fetch --all` and `git pull origin main`
- Switch to the agent branch: `git checkout dev_[project_name]` (create from `main` if missing)
- Merge the fresh updates from main into the dev branch to prevent code drift: `git merge main`

### 2. Briefing
- read project `CLAUDE.md`
- Inspect `TODO.md` to bring the agent up to speed.
- Explicitly confirm the target task ID (e.g., `01`, `02`) with the user before performing any file operations.

### 4. Planning
- discuss with user if more planning is necessary
- discuss improvements, refinments

### 5. Execute
- Perform code modifications strictly within the local project directory on the `dev_[project_name]` branch.
- Adhere strictly to the guidelines specified during the Briefing phase.

### 6. Offboarding
- Count existing commits on the current branch to determine the next index number $N$ (via `git log --oneline`).
- Create a local commit with the strict format: `claude_[project_name] #[N]`
- Push the branch to remote: `git push origin claude_[project_name]`
- **JSON Dropzone Update:** Access the central log file absolutely located at `/home/flo/01_ai_stack/state_memory/projects_log.json`. 
  - Do not overwrite or delete previous historical session blocks.
  - Append a new, timestamped session block as a unique entry under the main keys to ensure a linear history ledger.
  - Under the block (`claude`): Update only `last_touch` and `status` inside the `metadata` object (keep `project_id` and `project_type` as immutable constants). Push new insights into the `abstract` and `manifesto` logs.
  - Under the task block (`todo`): Update the active `tasks` list with current states. Log the session changes chronologically into the history block.
  - Maintain a silent execution profile for this file write operation with minimal terminal output.
- **Safe Guard Checkout:** Right before closing the session, the agent MUST switch the working directory back to the main branch: `git checkout main`. This guarantees the user never accidentally types or works within the agent's development branch outside of Claude Code.

## git_guardrails
- **Branch Isolation:** Working directly on or committing to `main` or `master` is strictly forbidden.
- **Enforcement:** The agent must never automatically merge `dev_[project_name]` into `main`. Merging and deployment via Pull Requests are strictly reserved for the user.
- **Divergence Handling:** If any `git pull` or `git merge` fails due to conflicts, the agent must immediately STOP, display the exact git status, and hand complete control over to the user.

```python
import subprocess
from pathlib import Path

# --- HIER DEINE DATEINAMEN EINSTELLEN ---
source_notebook = "leleka_core_ctx.ipynb"  # Der Name deiner Datei
target_filename = "CLAUDE"        # Der Name der Zieldatei (ohne .md)
# ----------------------------------------

# Zielpfad im context-Ordner
target_folder = Path("/home/flo")

# Export-Befehl via Jupytext
# Er nimmt die Quelle und schreibt sie explizit in den Zielpfad mit dem neuen Namen
cmd = [
    "jupytext",
    "--to", "markdown",
    "--output", str(target_folder / f"{target_filename}.md"),
    source_notebook
]

try:
    subprocess.run(cmd, check=True)
    print(f"✓ Export erfolgreich! Datei '{source_notebook}' wurde als '{target_filename}.md' exportiert.")
except subprocess.CalledProcessError as e:
    print(f"✗ Fehler beim Export: {e}")
```
