```python
import subprocess
from pathlib import Path

# --- HIER DEINE DATEINAMEN EINSTELLEN ---
source_notebook = "leleka_core_ctx.ipynb"  # Der Name deiner Datei
target_filename = "CLAUDE"        # Der Name der Zieldatei (ohne .md)
# ----------------------------------------

# Zielpfad im context-Ordner
target_folder = Path("/home/flo")

# Export-Befehl via Jupytext
# Er nimmt die Quelle und schreibt sie explizit in den Zielpfad mit dem neuen Namen
cmd = [
    "jupytext",
    "--to", "markdown",
    "--output", str(target_folder / f"{target_filename}.md"),
    source_notebook
]

try:
    subprocess.run(cmd, check=True)
    print(f"✓ Export erfolgreich! Datei '{source_notebook}' wurde als '{target_filename}.md' exportiert.")
except subprocess.CalledProcessError as e:
    print(f"✗ Fehler beim Export: {e}")
```
