```python
import json
import re
from datetime import datetime
from pathlib import Path

# Die zwei dedizierten Zielverzeichnisse für deine Exporte
OLD_EXPORT_DIR = Path("/home/flo/01_ai_stack/context/projects")  # Pfad aus image_1a2524.png
BASE_DEV_DIR = Path("/home/flo/02_dev")                                              # Pfad für die Dev-Umgebung

def run_export(notebook_path="spiderweb_projects.ipynb"):
    nb_file = Path(notebook_path)
    if not nb_file.exists():
        print(f"Fehler: Datei '{notebook_path}' nicht gefunden!")
        return

    with open(nb_file, "r", encoding="utf-8") as f:
        nb_data = json.load(f)

    today_str = datetime.now().strftime("%Y-%m-%d")
    export_count = 0

    for cell in nb_data.get("cells", []):
        tags = cell.get("metadata", {}).get("tags", [])

        # 1. Typ-Tag bestimmen (CLAUDE oder TODO)
        type_tag = next((t for t in tags if t in ["CLAUDE", "TODO"]), None)

        # 2. Projekt-Tag bestimmen (das jeweils andere Tag, das kein Typ ist)
        project_tag = next((t for t in tags if t not in ["CLAUDE", "TODO"]), None)

        if project_tag and type_tag:
            # --- INHALT VORBEREITEN ---
            raw_content = "".join(cell.get("source", []))
            last_touch_pattern = r"- \*\*last_touch:\*\*.*"
            new_last_touch_str = f"- **last_touch:** {today_str}"
            clean_raw = re.sub(last_touch_pattern, "", raw_content).strip()

            if re.search(last_touch_pattern, raw_content):
                updated_content = re.sub(last_touch_pattern, new_last_touch_str, raw_content)
            else:
                separator = "\n" if raw_content and not raw_content.endswith("\n") else ""
                updated_content = f"{raw_content}{separator}{new_last_touch_str}\n"

            # Helper-Funktion für den intelligenten Änderungs-Check
            def write_if_changed(file_path, content, clean_content):
                if file_path.exists():
                    existing = file_path.read_text(encoding="utf-8")
                    if re.sub(last_touch_pattern, "", existing).strip() == clean_content:
                        return False  # Keine Änderung im eigentlichen Text

                file_path.parent.mkdir(parents=True, exist_ok=True)
                with file_path.open("w", encoding="utf-8") as out:
                    out.write(content)
                print(f"Exportiert: {file_path}")
                return True

            # --- TARGET 1: Alter Kontext-Export (Flache .md Files im Context-Ordner) ---
            old_filename = f"project_{project_tag}.md"
            old_file_path = OLD_EXPORT_DIR / old_filename

            # --- TARGET 2: Neuer Dev-Export (Getrennte CLAUDE.md / TODO.md im dev-Ordner) ---
            new_target_dir = BASE_DEV_DIR / project_tag
            new_filename = f"{type_tag.upper()}.md"
            new_file_path = new_target_dir / new_filename

            # Beide Exporte unabhängig voneinander ausführen
            changed_old = write_if_changed(old_file_path, updated_content, clean_raw)
            changed_new = write_if_changed(new_file_path, updated_content, clean_raw)

            if changed_old or changed_new:
                export_count += 1

    print(f"Fertig! {export_count} Zellen wurden verarbeitet und abgeglichen.")

if __name__ == "__main__":
    run_export()
```
