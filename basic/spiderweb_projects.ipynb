```python
import json
import re
from datetime import datetime
from pathlib import Path

# Der einzige globale Ankerpunkt für deine Entwicklungsumgebung
BASE_DEV_DIR = Path("/home/flo/02_dev")

def run_export(notebook_path="spiderweb_projects.ipynb"):
    nb_file = Path(notebook_path)
    if not nb_file.exists():
        print(f"Fehler: Datei '{notebook_path}' nicht gefunden!")
        return

    with open(nb_file, "r", encoding="utf-8") as f:
        nb_data = json.load(f)

    today_str = datetime.now().strftime("%Y-%m-%d")
    export_count = 0

    for cell in nb_data.get("cells", []):
        tags = cell.get("metadata", {}).get("tags", [])

        # 1. Typ-Tag bestimmen (CLAUDE oder TODO)
        type_tag = next((t for t in tags if t in ["CLAUDE", "TODO"]), None)

        # 2. Projekt-Tag bestimmen (das jeweils andere Tag, das kein Typ ist)
        project_tag = next((t for t in tags if t not in ["CLAUDE", "TODO"]), None)

        # Nur verarbeiten, wenn beide Tags vorhanden sind
        if project_tag and type_tag:
            # Pfad: BASE_DEV_DIR / Projektname
            target_dir = BASE_DEV_DIR / project_tag
            # Dateiname: exakt CLAUDE.md oder TODO.md
            filename = f"{type_tag.upper()}.md"
            file_path = target_dir / filename

            # Rohen Inhalt extrahieren
            raw_content = "".join(cell.get("source", []))

            # Intelligenter Änderungs-Check
            has_changed = True
            if file_path.exists():
                existing_content = file_path.read_text(encoding="utf-8")

                clean_existing = re.sub(r"- \*\*last_touch:\*\*.*", "", existing_content)
                clean_raw = re.sub(r"- \*\*last_touch:\*\*.*", "", raw_content)

                if clean_existing.strip() == clean_raw.strip():
                    has_changed = False

            if has_changed:
                updated_content = re.sub(
                    r"- \*\*last_touch:\*\*.*",
                    f"- **last_touch:** {today_str}",
                    raw_content
                )

                target_dir.mkdir(parents=True, exist_ok=True)
                with file_path.open("w", encoding="utf-8") as out:
                    out.write(updated_content)

                print(f"Aktualisiert & Exportiert: {file_path}")
                export_count += 1

    print(f"Fertig! {export_count} geänderte Dateien wurden exportiert.")

if __name__ == "__main__":
    run_export()
```
