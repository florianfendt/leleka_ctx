# standards.md

## 01 conventions

### 01.01 general file rules
- **Naming Syntax:** All files and folders must be `lowercase`, using `snake_case`, and strictly **no spaces**.
- **Dates:** Use `YYYY-MM-DD` (e.g., `log_2026-03-25.md`).
- **Versioning:** No version numbers in filenames (handled via git/internal headers).

### 01.02 specific formats
- **Text & Documentation:** Use `.md` (Markdown). UTF-8 encoding.
- **Structured Data:** Use `.csv`. Separator: Semicolon (`;`). Headers in `snake_case`.
- **Logic & Scripts:** Use `.py` (Python). Follow PEP 8. Use `docstrings`.
- **Web Development:** `.html` (semantic tags) and `.css` (kebab-case for classes).
- **Data Exchange:** `.json`. Keys in `snake_case`. Indent: 2.

## 02 standards

### 02.01 markdown_standards
- **Line 1:** Must be H1 with exact filename: `# filename.extension`.
- **Level 2 Headlines:** Must be numbered: `## 01 [Title]`, `## 02 [Title]`.
- **Level 3 Headlines:** No numbering required.
- **Tables:** Mandatory for lists with >3 attributes per item.
- **Relational Logic:** Use `->` for flow and `decision:` for binding results.

### 02.02 python_standards
- **Style Guide:** Follow **PEP 8**.
- **Naming:** `snake_case` for variables, functions, and files.
- **Paths:** Use `pathlib` for OS-independent path handling. Use relative paths.
- **Documentation:** Every function needs a short `docstring` (Summary, Args, Returns).
- **Structure:** `if __name__ == "__main__":` block for executable scripts.

### 02.03 web_standards
- **HTML:** Use **HTML5 semantic tags** (`<main>`, `<section>`, `<nav>`).
- **CSS:**
  - Use **kebab-case** for classes and IDs (e.g., `.nav-bar`).
  - Use variables (`:root`) for colors/spacing to ensure consistency.
- **Separation:** Keep HTML, CSS, and JS in separate files.

### 02.04 json_standards
- **Keys:** Use `snake_case` for all keys.
- **Formatting:** Use `indent=2` for human readability.
- **Types:** Strict typing; no mixing of types in lists.

### 02.05 language standards
- **System Language:** All file content (Logic, Docs, Comments) must be in **ENGLISH**.
- **Interaction Language:** Chat and "Chain of Thought" must be in **GERMAN**.
