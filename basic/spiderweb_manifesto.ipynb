### network_devices

```python
import json
from pathlib import Path
import pandas as pd
from IPython.display import display, Markdown

# 1. Pfad zu deiner JSON-Datei
json_path = Path("/home/flo/00_system/config/network_devices.json")

try:
    # 2. Datei einlesen und JSON parsen
    with open(json_path, "r", encoding="utf-8") as f:
        data = json.load(f)
    # 3. In Pandas DataFrame konvertieren
    df = pd.DataFrame(data)

    # 4. Spalten sortieren, damit die wichtigsten Specs direkt vorne stehen
    preferred_order = ["id", "brand", "model", "chip", "ram", "os", "state", "type", "gpu"]
    existing_cols = [col for col in preferred_order if col in df.columns]
    remaining_cols = [col for col in df.columns if col not in existing_cols]

    df_ordered = df[existing_cols + remaining_cols]

    # =====================================================================
    # CSV-Export (Bleibt UNVERÄNDERT - schreibt alle Spalten)
    # =====================================================================
    output_csv_path = Path("/home/flo/01_ai_stack/context/system/network_infrastructure.csv")

    # Sicherstellen, dass die Ordnerstruktur existiert
    output_csv_path.parent.mkdir(parents=True, exist_ok=True)

    # Speichern als CSV ohne den Pandas-Index
    df_ordered.to_csv(output_csv_path, index=False, sep=",", encoding="utf-8")
    #  print(f"CSV erfolgreich exportiert nach: {output_csv_path}")

    # =====================================================================
    # 5. Schöne Tabelle ausgeben (GEÄNDERT - zeigt nur bestimmte Spalten)## 1. Workspace & Workflow
    # =====================================================================

    # Hier sind die gefilterten Spalten (anhand deiner Sprachnachricht interpretiert).
    # Falls eine Spalte in der JSON anders heißt (z.B. "network" statt "net"),
    # passe den Namen hier in der Liste einfach an:
    display_columns = ["id", "brand", "model", "chip", "ram", "os", "type", "disk_os", "role", "net"]

    # Wir filtern die Liste sicherheitshalber, damit es keinen Fehler gibt,
    # falls eine Spalte in der JSON-Datei fehlt:
    available_display_cols = [col for col in display_columns if col in df_ordered.columns]

    # Neues (kleineres) DataFrame nur für die Anzeige erstellen
    df_display = df_ordered[available_display_cols]

    # Markdown-Tabelle rendern
    print(df_display.to_csv(index=False, sep="\t"))

except FileNotFoundError:
    print(f"Fehler: Die Datei unter '{json_path}' wurde nicht gefunden. Bitte passe den Pfad an.")
except json.JSONDecodeError:
    print("Fehler: Die Datei enthält kein gültiges JSON-Format.")
```

### workspace_tree

```python
import json
from pathlib import Path

def display_system_tree(json_file="/home/flo/00_system/config/paths.json", mode="core"):
    """
    Gibt die Pfade token-optimiert als gruppierten Text aus.
    mode="core": Minimaler Kontext (nur Basis-Ordner, weniger Beschreibungen).
    mode="main": Voller Kontext (alle Dirs und Files inkl. Beschreibungen).
    """
    with open(json_file, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # Definiert, was im 'core'-Modus zwingend für den Basis-Agenten nötig ist
    core_aliases = {"documents", "pictures", "videos", "dropzone", "context", "pulse"}
    # Standard-Ordner, bei denen die Beschreibung im 'core' gelöscht wird, um Token zu sparen
    self_explanatory = {"documents", "pictures", "videos"}

    print(f"--- Workspace Tree ({mode.upper()}) ---")

    if mode == "main":
        print("\n[DIRECTORIES]")
        for alias, info in data.get('paths', {}).items():
            desc = f" - {info['desc']}" if info.get('desc') else ""
            print(f"* {alias}: {info['path']}{desc}")

        print("\n[FILES]")
        for name, info in data.get('targets', {}).items():
            desc = f" - {info['desc']}" if info.get('desc') else ""
            print(f"* {name}: {info['path']}{desc}")

    elif mode == "core":
        print("\n[CORE STORAGE & PIPELINE]")
        for alias, info in data.get('paths', {}).items():
            if alias in core_aliases:
                # Beschreibung nur dranhängen, wenn der Ordner nicht selbsterklärend ist
                desc = ""
                if info.get('desc') and alias not in self_explanatory:
                    desc = f" - {info['desc']}"
                print(f"* {alias}: {info['path']}{desc}")

if __name__ == "__main__":
    # Test-Aufruf für den minimalen Kontext
    display_system_tree(mode="main")

    # Für vollen Output später einfach so aufrufen:
    # display_system_tree(mode="main")
```
